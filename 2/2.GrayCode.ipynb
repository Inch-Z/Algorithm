```python
import os
import time
PATH='.\\'
```

```python
class Gray():
    def calCode(self, n:int) -> list:#核心算法
        return [0] if n == 0 else self.calCode(n-1) + [x+(1<<(n-1)) for x in reversed(self.calCode(n-1))]
    def calCode_1(self, n:int) -> list:#解释calCode
        if n == 0:
            result = [0]
        else:
            temp=[]
            for x in reversed(self.calCode_1(n-1)):
                temp.append(x+(1<<(n-1)))
            result = self.calCode_1(n-1)+temp
        return result
    def calCode_2(self, n:int) -> list:#最优算法 利用格雷码电路生成原理
        return [i^(i>>1) for i in range(2**n)]
```

```python
start = time.clock()
n=eval(input("请输入格雷码位数:\n"))
ans=Gray()
result=[]
b=ans.calCode(n)
for i in b:
    result.append(bin(i)[2:].zfill(n))
print(result)
end = time.clock()
print (end-start)
##文件输出
f = open(PATH + '2.GrayCode_result', 'w', encoding='UTF-8')
for i in range(2 ** n):
    f.write(result[i])
    f.write('\n')
f.close()
```
